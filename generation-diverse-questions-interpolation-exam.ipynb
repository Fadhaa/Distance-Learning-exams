**This Notebook shows how to generate divers exam about a topic of interpolation by use five methods, namely Newton Forward, Newton Backward, Newton for divided differences, Gauss Forward, Gauss Backward**

```r
# define the switch function
mm<- function(A, type){
switch(type,
polf=polyf(A[1,],A[2,]),
polb=polyb(A[1,],A[2,]),
pold=polyd(A[1,],A[2,]),
gaussf=gaussf(A[1,],A[2,]),
gaussb=gaussb(A[1,],A[2,]))
}
```

**function for interpolation of points by Gauss Forward Method**

```r
gaussf<-function(xi,yi){
li<-c()
L<-length(xi)
x0<-xi[3]
h<-xi[2]-xi[1]
x<-xi[L]
k<-1
yy<-yi
d<-rep(0,5)
a<-rep(0,5)
dd<-c()
d[k]<-yy[3]
while (length(yy)>1){
for (i in 1:(length(yy)-1)){
dd<-cbind(dd,yy[i+1]-yy[i])
}
li<-c(li,dd)
k<-k+1
if (k==2){
r=3}else if (k==3){
r=2
}else if (k==4){
r=2}else if (k==5){
r=1}
d[k]<-dd[r]
yy<-dd
dd<-c()
}
valu<-c(li[1],li[2],li[3],li[4],li[5],li[6],li[7],li[8],li[9],li[10])
v<-d
a[1]=v[5]/(24*h^4)
a[2]=(v[4]/(6*h^3))-((4*v[5])/(24*h^4))*x0-((2*v[5])/(24*h^4))*h
a[3]=(v[3]/(2*h^2))-((3*v[4])/(6*h^3))*x0-((v[5])/(24*h^4))*h^2+((6*v[5])/(24*h^4))*x0*h+((6*v[5])/(24*h^4))*x0^2
a[4]=(v[2]/h)-(2*v[3]/(2*h^2))*x0-(v[3]/(2*h^2))*h+((3*v[4])/(6*h^3))*x0^2-((v[4])/(6*h^3))*h^2-(3*v[5]/(24*h^4))*x0^3-((v[5])/(24*h^4))*x0^3+((v[5])/(24*h^4))*x0*h^2-((6*v[5])/(24*h^4))*h*x0^2+((v[5])/(24*h^4))*x0*h^2+((2*v[5])/(24*h^4))*h^3
a[5]=v[1]-(v[2]/h)*x0+(v[3]/(2*h^2))*x0^2+(v[3]/(2*h^2))*h*x0-(v[4]/(6*h^3))*x0^3+(v[4]/(6*h^3))*x0*h^2-(v[5]/(24*h^4))*h^2*x0^2+(2*v[5]/(24*h^4))*h*x0^3-(2*v[5]/(24*h^4))*h^3*x0+(v[5]/(24*h^4))*x0^4
px<-a[1]*x^4+a[2]*x^3+a[3]*x^2+a[4]*x+a[5]
lis<-list(d=valu,a=a,px=px)
return(lis)}
```

**function for interpolation of points by Gauss Backward Method**

```r
gaussb<-function(xi,yi){
li<-c()
L<-length(xi)
x0<-xi[3]
h<-xi[2]-xi[1]
x<-xi[L]
k<-1
yy<-yi
d<-rep(0,5)
a<-rep(0,5)
dd<-c()
d[k]<-yy[3]
while (length(yy)>1){
for (i in 1:(length(yy)-1)){
dd<-cbind(dd,yy[i+1]-yy[i])
}
li<-c(li,dd)
k<-k+1
if (k==2){
r=2}else if (k==3){
r=2
}else if (k==4){
r=1}else if (k==5){
r=1}
d[k]<-dd[r]
yy<-dd
dd<-c()
}
valu<-c(li[1],li[2],li[3],li[4],li[5],li[6],li[7],li[8],li[9],li[10])
v<-d
a[1]=v[5]/(24*h^4);
a[2]=(v[4]/(6*h^3))-((4*v[5])/(24*h^4))*x0+((2*v[5])/(24*h^4))*h
a[3]=(v[3]/(2*h^2))-((3*v[4])/(6*h^3))*x0-((v[5])/(24*h^4))*h^2-((6*v[5])/(24*h^4))*x0*h+((6*v[5])/(24*h^4))*x0^2
a[4]=(v[2]/h)-(2*v[3]/(2*h^2))*x0+(v[3]/(2*h^2))*h+((3*v[4])/(6*h^3))*x0^2-((v[4])/(6*h^3))*h^2-(3*v[5]/(24*h^4))*x0^3-((v[5])/(24*h^4))*x0^3+((v[5])/(24*h^4))*x0*h^2+((6*v[5])/(24*h^4))*h*x0^2+((v[5])/(24*h^4))*x0*h^2-((2*v[5])/(24*h^4))*h^3
a[5]=v[1]-(v[2]/h)*x0+(v[3]/(2*h^2))*x0^2-(v[3]/(2*h^2))*h*x0-(v[4]/(6*h^3))*x0^3+(v[4]/(6*h^3))*x0*h^2-(v[5]/(24*h^4))*h^2*x0^2-(2*v[5]/(24*h^4))*h*x0^3+(2*v[5]/(24*h^4))*h^3*x0+(v[5]/(24*h^4))*x0^4
px<-a[1]*x^4+a[2]*x^3+a[3]*x^2+a[4]*x+a[5]
lis<-list(d=valu,a=a,px=px)
return(lis)}
```

**function for interpolation of points by Newton Forward Method**

```r
polyf<-function(xi,yi){
li<-c()
L<-length(xi)
x0<-xi[1]
h<-xi[2]-xi[1]
x<-xi[L]
k<-1
yy<-yi
d<-rep(0,5)
a<-rep(0,5)
dd<-c()
d[k]<-yy[1]
while (length(yy)>1){
for (i in 1:(length(yy)-1)){
dd<-cbind(dd,yy[i+1]-yy[i])
}
li<-c(li,dd)
k<-k+1
d[k]<-dd[1]
yy<-dd
dd<-c()
}
valu<-c(li[1],li[2],li[3],li[4],li[5],li[6],li[7],li[8],li[9],li[10])
v<-d
a[1]=v[5]/(24*h^4)
a[2]=(v[4]/(6*h^3))-((4*v[5])/(24*h^4))*x0-((6*v[5])/(24*h^4))*h
a[3]=(v[3]/(2*h^2))-((3*v[4])/(6*h^3))*x0-((3*h*v[4])/(6*h^3))+((6*v[5])/(24*h^4))*x0^2+((18*v[5])/(24*h^4))*x0*h+((11*v[5])/(24*h^4))*h^2
a[4]=(v[2]/h)-(v[3]/(h^2))*x0-(v[3]/(2*h^2))*h+((2*v[4])/(6*h^3))*x0^2+((6*v[4])/(6*h^3))*h*x0+(v[4]/(6*h^3))*x0^2+((2*v[4])/(6*h^3))*h^2-((22*v[5])/(24*h^4))*x0*h^2-((18*v[5])/(24*h^4))*h*x0^2-((4*v[5])/(24*h^4))*x0^3-((6*v[5])/(24*h^4))*h^3
a[5]=v[1]-(v[2]/h)*x0+(v[3]/(2*h^2))*x0^2+(v[3]/(2*h^2))*h*x0+((v[4]/(6*h^3))*(-3*h*x0^2-2*h^2*x0-x0^3))+((v[5]/(24*h^4))*(11*x0^2*h^2+6*h*x0^3+6*x0*h^3+x0^4))
px<-a[1]*x^4+a[2]*x^3+a[3]*x^2+a[4]*x+a[5]
lis<-list(d=valu,a=a,px=px)
return(lis)}
```

**function for interpolation of points by Newton Backward Method**

```r
polyb<-function(xi,yi){
li<-c()
L<-length(xi)
x0<-xi[L]
h<-xi[2]-xi[1]
x<-xi[L]
k<-1
yy<-yi
d<-rep(0,5)
a<-rep(0,5)
dd<-c()
d[k]<-yy[L]
while (length(yy)>1){
for (i in 1:(length(yy)-1)){
dd<-cbind(dd,yy[i+1]-yy[i])
}
li<-c(li,dd)
k<-k+1
Lk<-length(dd)
d[k]<-dd[Lk]
yy<-dd
dd<-c()
}
valu<-c(li[1],li[2],li[3],li[4],li[5],li[6],li[7],li[8],li[9],li[10])
v<-d
a[1]=v[5]/(24*h^4)
a[2]=((v[4]/(6*h^3))-((4*v[5])/(24*h^4))*x0+((6*v[5])/(24*h^4))*h)
a[3]=(v[3]/(2*h^2))-((3*v[4])/(6*h^3))*x0+((3*v[4])/(6*h^3))*h+((6*v[5])/(24*h^4))*x0^2-((18*v[5])/(24*h^4))*x0*h+((11*v[5])/(24*h^4))*h^2
a[4]=(v[2]/h)-(v[3]/h^2)*x0+(v[3]/(2*h^2))*h+((2*v[4])/(6*h^3))*x0^2-((6*v[4])/(6*h^3))*h*x0+(v[4]/(6*h^3))*x0^2+((2*v[4])/(6*h^3))*h^2-((22*v[5])/(24*h^4))*x0*h^2+((18*v[5])/(24*h^4))*h*x0^2-((4*v[5])/(24*h^4))*x0^3+((6*v[5])/(24*h^4))*h^3
a[5]=v[1]-(v[2]/h)*x0+(v[3]/(2*h^2))*x0^2-(v[3]/(2*h^2))*h*x0+((v[4]/(6*h^3))*(+3*h*x0^2-2*h^2*x0-x0^3))+((v[5]/(24*h^4))*(11*x0^2*h^2-6*h*x0^3-6*x0*h^3+x0^4));
px<-a[1]*x^4+a[2]*x^3+a[3]*x^2+a[4]*x+a[5]
lis<-list(d=valu,a=a,px=px)
return(lis)}
```

**function for interpolation of points by Newton for divided differences Method**

```r
polyd<-function(xi,yi){
li<-c()
L<-length(xi)
y<-xi
x<-xi[5]
k<-1
d<-rep(0,5)
a<-rep(0,5)
dd<-c()
yy<-yi
d[k]<-yy[1]
while (length(yy)>1){
for (i in 1:(length(yy)-1)){
dd<-cbind(dd,(yy[i+1]-yy[i])/(y[i+k]-y[i]))
}
li<-c(li,dd)
k<-k+1
d[k]<-dd[1]
yy<-dd
dd<-c()
}
valu<-c(li[1],li[2],li[3],li[4],li[5],li[6],li[7],li[8],li[9],li[10])
v<-d
a[1]<-v[5]
a[2]<-v[4]-(v[5]*y[1])-(v[5]*y[2])-(v[5]*y[3])-(v[5]*y[4])
a[3]<-v[3]-(v[4]*y[1])-(v[4]*y[2])-(v[4]*y[3])+(v[5]*y[2]*y[1])+(v[5]*y[3]*y[1])+(v[5]*y[3]*y[2])+(v[5]*y[4]*y[1])+(v[5]*y[4]*y[2])+(v[5]*y[4]*y[3])
a[4]<-v[2]-(v[3]*y[1])-(v[3]*y[2])+(v[4]*y[2]*y[1])+(v[4]*y[3]*y[1])+(v[4]*y[3]*y[2])-(v[5]*y[3]*y[2]*y[1])-(v[5]*y[4]*y[2]*y[1])-(v[5]*y[4]*y[3]*y[1])-(v[5]*y[4]*y[3]*y[2])
a[5]<-v[1]-(v[2]*y[1])+(v[3]*y[2]*y[1])-(v[4]*y[3]*y[2]*y[1])+(v[5]*y[4]*y[3]*y[2]*y[1])
px<-a[1]*x^4+a[2]*x^3+a[3]*x^2+a[4]*x+a[5]
lis<-list(d=valu,a=a,px=px)
return(lis)}
```
